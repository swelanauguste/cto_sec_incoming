```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('./app/incoming.csv')
```

```python
df
```

```python
df['received_on_str'] = pd.to_datetime(df['received_on'], format='%m/%d/%Y', errors='coerce')
```

```python
df
```

```python
df['received_on'] = df['received_on_str'].dt.strftime('%Y-%m-%d')
```

```python
df.to_csv('data.csv', sep=',', encoding='utf-8', index=False, header=True)
```
